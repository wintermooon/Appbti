# 수요는 높지만 만족도가 평균치보다 낮아 도전해 볼만한 시장 분석하기 

### 수요 판단하기 
- 수요가 높다고 판단하는 경우: 전체 앱을 카테고리별로 묶었을 때 순위 와 100만 다운로드 이상인 앱을 카테고리 별로 묶었을 떄 순위가 2단계 이상 올라간다면 수요 높음 (비율의 퍼센트도 봐야하는 가?)
- 수요가 낮다고 판단하는 경우: 전체 앱을 카테고리 별로 묶었을 때 순위 와 100만 다운로드 이상인 앱을 카테고리 별로 묶었을 때 순위가 2단계 이상 내려간다면 수요 낮음

- 수요와 공급이 일치하는 경우: 나머지의 경우 

### 만족도가 평균치보다 낮아 도전해 볼만한 시장인지 판단하기

- 각 카테고리 별 평점을 봤을 때, 3점대 어플이 더 많은 카테고리: 예를 들어 게임 카테고리만 평점별 앱 개수를 봤을 때, 3점대가 많다면 만족도가 평균치보다 낮아 도전해볼 만한 시장이라고 판단 













```python
# 카테고리 별 묶었을 때 앱의 개수 순위
category_app = data.groupby('Category').size().reset_index(name = 'counts_total').sort_values('Category').reset_index(drop = True)
category_total = category_app[['Category']].copy()
category_total['Total Rank'] = category_app['counts_total'].rank(method='min', ascending = False)
category_total
```

```python
# 카테고리 별 묶었을 때 앱의 개수 순위
category_100M = pd.DataFrame(data[data["Maximum Installs"] >= 1000000].groupby('Category').size().reset_index(name = 'counts_100M'))

category_rank_100M = category_100M[['Category']].copy()
category_rank_100M['100M Installs Rank'] = category_100M['counts_100M'].rank(method='min', ascending = False)
category_rank_100M 

```

```python
#  카테고리 별 순위
category_rank = pd.merge(category_total, category_rank_100M).sort_values('Total Rank')

category_rank
```

# 예시 
1. 100만 이상 다운로드 & 평점 3점대인 어플의 카테고리 비율 4위는 Photography 가 나왔습니다.   <br/>

이를 분석하면, 

<br/>

첫째, 전체 앱 카테고리 에서 Photography 카테고리는 11위로 낮은 편입니다.<br/>   이를 통해 출시되는 앱의 수가 적은편이라는 것을 알 수 있습니다. <br/> 

반면, 100만 다운로드 이상되는 앱 중에서 Photography 카테고리는 3위로 높은 편입니다.<br/> 이를통해, Photography app 의 수요가 높은 편이라고 볼 수있습니다. 

<br/>

둘째, 만족도가 최빈값보다 낮은 3점대 어플 중 Photography 카테고리의 순위는 8위 입니다. <br/> 반면, 100만 다운로드 이상이면서, 평점 3점대 인 앱 중 Photography 카테고리는 4위입니다. <br/> 
이를통해, 만족도가 최빈값보다 약간 낮은 평점 3점대 어플에서 Photography 카테고리가 수요가높다는 것을 알 수 있습니다. 
<br/>

즉, Photography: 카테고리가 수요는 높지만, 만족도가 평균치 보다 낮아 도전해 볼만한 시장이라 볼 수 있을 것 같습니다. 

```python
# 3점대 카테고별 앱 개수
rating3_apps =data[data['IntRating'] == 3].groupby('Category').size()

#  카테고리 별 전체 앱 개수
total_apps = data.groupby('Category').size()

# 카테고리 내에서 3점대 어플의 비율 과 그 순위
ratio_rating3 = pd.DataFrame((rating3_apps/total_apps)*100)
ratio_rating3.columns = ['Ratio']
ratio_rating3['Ratio Rank'] = ratio_rating3['Ratio'].rank(ascending =False)
ratio_rating3['Ratio Rank'] = ratio_rating3['Ratio Rank'].astype(int)
new = ratio_rating3.copy()
new.sort_values('Ratio Rank')
```
