```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# notebook을 실행한 브라우저에서 바로 그림을 볼 수있게 해주는 명령어
# matplotlib 시각화 결과를 jupyter notebook에서 바로 확인하기 위한 코드 작성 
%matplotlib inline
```

```python
# csv 파일 불러오기
# 파일 경로는 colab이용시 드라이브 경로, vsc이용시 로컬경로
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/google-app-pre.csv')
```

```python
# 제대로 불러와졌는지 확인하고싶으면
print(data.shape) # 행, 열의 개수 확인 가능
```

```python
data
```

```python
#광고 있는 어플들 평점 순서대로
AdTrue =  data[data["Ad Supported"]== True].sort_values('Rating', ascending = False).reset_index(drop =False)
```

```python
AdTrue
```

```python
AdTrue.to_json('/content/drive/MyDrive/Colab Notebooks/data/AdTrue.json')
```

```python
#광고 있는 것 중 평점 5점짜리
AdTrue5 = AdTrue[AdTrue["Rating"] == 5].sort_values('Maximum Installs', ascending = False)
```

```python
AdTrue5.to_json('/content/drive/MyDrive/Colab Notebooks/data/AdTrue5.json')
```

```python
AdTrue5
```

```python
#광고 있는 것 중 평점 5점짜리 카테고리 그룹화
ctg_ad5 = AdTrue5[data['Ad Supported'] == True].groupby('Category').count().sort_values('Category', ascending = False).reset_index(drop =False)
```

```python
ctg_ad5.to_json('/content/drive/MyDrive/Colab Notebooks/data/ctg_ad5.json')
```

```python
ctg_ad5
```

```python
#광고+5점 카테고리 데이터프레임 정리(인덱스만 담기)
ctg_ad5_c = ctg_ad5.drop(['Unnamed: 0', 'App Name',	'App Id',	'Rating',	'Maximum Installs',	'Free',	'Price',	'Currency',	'Released',	'Last Updated',	'Content Rating',	'Ad Supported',	'In App Purchases',	'Editors Choice'], axis=1).sort_values(by='index', ascending = False).reset_index(drop =True)
```

```python
ctg_ad5_c.to_json('/content/drive/MyDrive/Colab Notebooks/data/ctg_ad5_c.json')
```

```python
ctg_ad5_c
```

```python
#광고 있고 평점 5점 다운로드수로 정렬
MI_ad5 = AdTrue5.drop(['Unnamed: 0', 'App Id',	'Rating',	'Free',	'Price', 'index', 'Currency',	'Released',	'Last Updated',	'Content Rating',	'Ad Supported',	'In App Purchases',	'Editors Choice'], axis=1).reset_index(drop =True)
```

```python
MI_ad5.to_json('/content/drive/MyDrive/Colab Notebooks/data/MI_ad5.json')
```

```python
#광고 있고 평점 5점 다운로드수 상위 10개 중 인앱 구매가 6개, 상위 20개 중에 인앱 구매가 7개
MI_ad5
```

```python
#광고 있고 평점 5점 인앱 구매 있음 87개
#게임이 23개로 압도적으로 많음
AdInAppT = AdTrue5[AdTrue5["In App Purchases"] == True].drop(['Unnamed: 0',	'App Id',	'Rating',		'Free',	'Price',	'Currency',	'Released',	'Last Updated',	'Content Rating',	'Ad Supported',	'In App Purchases',	'Editors Choice', 'index'], axis=1).sort_values(by='Maximum Installs', ascending = False).reset_index(drop = True)
```

```python
AdInAppT.to_json('/content/drive/MyDrive/Colab Notebooks/data/AdInAppT.json')
```

```python
AdInAppT
```

```python
#광고 있고 평점 5점 인앱 구매 없음 296개
AdInAppF = AdTrue5[AdTrue5["In App Purchases"] == False].drop(['Unnamed: 0',	'App Id',	'Rating',		'Free',	'Price',	'Currency',	'Released',	'Last Updated',	'Content Rating',	'Ad Supported',	'In App Purchases',	'Editors Choice', 'index'], axis=1).sort_values(by='Maximum Installs', ascending = False).reset_index(drop = True)
```

```python
AdInAppF.to_json('/content/drive/MyDrive/Colab Notebooks/data/AdInAppF.json')
```

```python
AdInAppF
```

```python
#인앱구매 있는거 카테고리 수
AdInAppTC= AdInAppT[data['Ad Supported'] == True].drop(['Maximum Installs'], axis=1).groupby('Category').count().sort_values('App Name', ascending = False).reset_index(drop =False)
```

```python
AdInAppTC.to_json('/content/drive/MyDrive/Colab Notebooks/data/AdInAppTC.json')
```

```python
#인앱구매 없는거 카테고리 수
AdInAppFC= AdInAppF[data['Ad Supported'] == True].drop(['Maximum Installs'], axis=1).groupby('Category').count().sort_values('App Name', ascending = False).reset_index(drop = False)
```
